```python
from typing import List


class Solution:
    def maxMSTWeight(self, n: int, m: int, strings: List[str]) -> int:
        if n == 1:
            return 0

        # 破环成链：翻倍拼接
        doubled = [s + s for s in strings]

        # 计算完全图的边权（最长环形公共子串长度）
        weights = [[0] * n for _ in range(n)]
        for i in range(n):
            for j in range(i + 1, n):
                lo, hi, best = 1, m, 0
                while lo <= hi:
                    mid = (lo + hi) // 2
                    # 检查 s_i 的任意环形子串是否出现在 s_j 的翻倍串中
                    found = any(doubled[i][k:k+mid] in doubled[j] for k in range(m))
                    if found:
                        best = mid
                        lo = mid + 1
                    else:
                        hi = mid - 1
                weights[i][j] = weights[j][i] = best

        # Prim 算法求最大生成树
        visited = [False] * n
        max_w = [weights[0][i] for i in range(n)]
        visited[0] = True
        total = 0

        for _ in range(n - 1):
            u = max((i for i in range(n) if not visited[i]), key=lambda i: max_w[i])
            visited[u] = True
            total += max_w[u]
            for i in range(n):
                if not visited[i] and weights[u][i] > max_w[i]:
                    max_w[i] = weights[u][i]

        return total

```

```python
# --- 测试 ---
sol = Solution()

# 示例测试
print(sol.maxMSTWeight(3, 4, ["abcd", "bcda", "cdab"]))  # 环形旋转相同，边权都是4
print(sol.maxMSTWeight(2, 3, ["abc", "xyz"]))             # 无公共子串，边权0
print(sol.maxMSTWeight(1, 5, ["hello"]))                  # 单节点，返回0
```
