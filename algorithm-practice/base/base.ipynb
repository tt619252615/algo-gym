```python
def calculate_final_score(daily: int, mid: int, final: int) -> float:
    """
    根据权重计算总评成绩。
    
    权重分配：平时(20%)，期中(30%)，期末(50%)。
    
    Args:
        daily: 平时成绩 (0-100)
        mid: 期中成绩 (0-100)
        final: 期末成绩 (0-100)
        
    Returns:
        float: 计算后的总评成绩
    """
    # 使用数值权重，保持逻辑清晰
    return daily * 0.2 + mid * 0.3 + final * 0.5

def validate_scores(*scores: int) -> bool:
    """
    校验所有输入成绩是否在 0-100 之间。
    
    Args:
        *scores: 变长参数，接收多个成绩整数
        
    Returns:
        bool: 全部合法返回 True，否则返回 False
    """
    return all(0 <= s <= 100 for s in scores)
 # 程序入口
try:
    # 模拟一行输入：例如输入 "80 90 85"
    raw_input = input("请输入平时、期中、期末成绩（空格分隔）：")
    score_list = [int(x) for x in raw_input.split()]
    
    if len(score_list) == 3 and validate_scores(*score_list):
        total = calculate_final_score(*score_list)
        print(f"总评成绩: {total:.2f}")
    else:
        print("成绩输入错误")
except ValueError:
    print("成绩输入错误")
```

```python
## 阶梯水费
def get_water_fee(usage: float) -> float:
    """
    使用分段逻辑计算阶梯水费。
    
    计费标准：
    - <= 10t: 2.5元/t
    - 10t < x <= 20t: 10*2.5 + (x-10)*3.5
    - > 20t: 10*2.5 + 10*3.5 + (x-20)*5.0
    
    Args:
        usage: 用水量 (float)
        
    Returns:
        float: 总费用
    """
    if usage <= 10:
        return usage * 2.5
    elif usage <= 20:
        return 10 * 2.5 + (usage - 10) * 3.5
    else:
        return 10 * 2.5 + 10 * 3.5 + (usage - 20) * 5.0
# 程序入口
try:
    user_input = input("请输入月用水量（吨）：")
    usage_val = float(user_input)
    
    if usage_val < 0:
        print("输入无效：用水量不能为负数")
    else:
        fee = get_water_fee(usage_val)
        print(f"应付水费:{fee:.2f}元")
except ValueError:
    print("输入无效：请输入数字")
```

```python
# 猜数字
import random

def play_game() -> None:
    """
    运行单轮猜数字游戏。
    生成1-100的随机数，通过循环接受用户输入并给出提示，直到猜中为止。
    """
    target_number = random.randint(1, 100)
    attempts = 0
    
    print("\n--- 新游戏开始！我已生成了一个 1-100 之间的整数 ---")
    
    while True:
        try:
            user_input = input("请输入你猜测的数字：").strip()
            guess = int(user_input)
            attempts += 1
            
            if guess > target_number:
                print("猜大了")
            elif guess < target_number:
                print("猜小了")
            else:
                print(f"恭喜你，猜对了! 共猜了 {attempts} 次")
                break  # 猜中了，跳出当前游戏的循环
                
        except ValueError:
            print("输入无效，请输入一个整数。")

def start_session() -> None:
    """
    管理游戏会话的主循环。
    负责询问用户是否继续游戏，并处理 y/n 的逻辑跳转。
    """
    # 首次启动游戏
    play_game()
    
    while True:
        choice = input("是否继续游戏？(y/n): ").strip().lower()
        
        if choice in ('y', 'yes'):
            play_game()
            continue  # 继续下一次询问循环
        elif choice in ('n', 'no'):
            print("感谢游玩，再见！")
            break     # 彻底退出主循环
        else:
            print("输入无效，请重新输入")
            # 此处不需要写 continue，循环会自动回到开头询问 choice
# 程序入口
start_session()
```

```python
### 找出指定范围素数
def is_prime(num: int) -> bool:
    """
    判断一个整数是否为素数。
    
    Args:
        num: 待检测的整数
        
    Returns:
        bool: 如果是素数返回 True，否则返回 False
    """
    if num < 2:
        return False
    # 只需检查到平方根即可，提高效率
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def find_primes_in_range(m: int, n: int) -> list[int]:
    """
    获取区间 [m, n] 内的所有素数。
    
    Args:
        m: 起始整数
        n: 结束整数
        
    Returns:
        list[int]: 素数列表
    """
    return [x for x in range(m, n + 1) if is_prime(x)]

def display_primes(primes: list[int], per_line: int = 5, width: int = 6) -> None:
    """
    按特定格式输出素数列表。
    
    Args:
        primes: 素数列表
        per_line: 每行显示的个数
        width: 每个数字占用的字符宽度
    """
    if not primes:
        print("该范围内没有素数")
        return

    for i, p in enumerate(primes):
        # :>{width} 表示右对齐并占用指定宽度
        print(f"{p:>{width}}", end="")
        
        # 每输出 per_line 个数字后换行
        if (i + 1) % per_line == 0:
            print()
            
    # 如果最后一行不满 per_line 个，手动补一个换行
    if len(primes) % per_line != 0:
        print()
try:
    # 模拟输入，例如输入 "10 50"
    raw_input = input("请输入两个正整数 m 和 n (空格分隔): ").strip()
    m, n = map(int, raw_input.split())
    
    if m >= n:
        print("输入错误：必须满足 m < n")
    else:
        # 1. 寻找素数
        prime_list = find_primes_in_range(m, n)
        
        # 2. 格式化输出
        display_primes(prime_list)

except ValueError:
    print("输入无效：请输入两个正整数")
```
